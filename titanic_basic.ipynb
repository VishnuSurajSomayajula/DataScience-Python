```python
import numpy as np
import pandas as pd
```

```python
titanic_data = pd.read_csv(r'E:\vs code\datasets\titanic.csv')
titanic_data.head(30)
```

```python
#in dataset , we selected survived Column and checked the unique values in it
uniquevalue_survived = titanic_data["Survived"].unique()
print("the unique values in 'Survived' is " ,uniquevalue_survived, sep= ' ')

uniquevalue_sex = titanic_data["Sex"].unique()
print("the unique values in 'Sex' is " ,uniquevalue_sex, sep= ' ')

# fr = ds["Fare"].unique()
# print("the unique values in 'Fare' is " ,fr, sep= ' ')
```

```python
newvalue_survived = pd.Categorical(titanic_data["Survived"])
renamevalue_survived = newvalue_survived.rename_categories(["dead","Survived"])
print(renamevalue_survived.describe())
```

```python
uniquevalue_Pclass = titanic_data["Pclass"].unique()
print("the unique values in 'Passenger Class' is " ,uniquevalue_Pclass, sep= ' ')

newvalue_Pclass  = pd.Categorical(titanic_data["Pclass"], ordered= True)
print(newvalue_Pclass.sort_values)
renamevalue_Pclass = newvalue_Pclass.rename_categories(["Class1","Class2","Class3"])
print(renamevalue_Pclass.describe())
```

```python
titanic_data['Pclass']= renamevalue_Pclass
titanic_data.head(10)
```

```python
uniquevalue_cabin = titanic_data["Cabin"].unique
print(np.array(uniquevalue_cabin))

charvalue_cabin = titanic_data["Cabin"].astype(str)
print(type(charvalue_cabin))
char_1stvalue_cabin = np.array([i[0] for i in charvalue_cabin])
char_1stvalue_cabin = pd.Categorical(char_1stvalue_cabin)
char_1stvalue_cabin.describe()
```

```python
titanic_data["Cabin"] = char_1stvalue_cabin
titanic_data.head()
```
